### Task 3

We can use a Monte Carlo approach to sample random points in a 3D space and then check if the points are blocked by DNA and in that way we can estimate the accessible volume. 

First we use the coordinates of the DNA atoms given in the dna_coords.txt file. Then we design a box which will be the "total" space. We are going to generate a certain amount of random points within that box. We also define a probe radius.

For each point we generate, we will check whether it is accessible or not, by computing the points distance to the DNA atoms, since each DNA atom occupies a sphere, the probe radius can not enter an area where it would overlap with an atom. And that area would be The probe radius plus the radius of the sphere which the DNA atom occupies:  $$ A_{occupied} = r_{probe} + r_{DNA} $$

So for each atom we will calculate the distance between the point and that atom, and if the distance is larger than the probe radius plus the radius of the DNA, that point is accessible.

We will then calculate the fraction of accessible points, by dividing the accsessible points we have checked, with the total amount of random generated points: $$ f_{points} = \frac{P_{accessible}}{P_{total}} $$

 We find the accsessible volume when we multiply the fraction of accessible points with the total volume of the box:

 $$ V_{accessible} = f_{points} \cdot V_{box} $$
